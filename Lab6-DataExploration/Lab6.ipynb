## Introduction 
the comparison of this data is to compare the average lifespan of people after diagnosis of Glioblastoma. a type of malignant tumor affecting the brain or spine. the disease itself is uncurable however methods to prolong the life of patients have been disccovered and have been shown to show promising results. this data was taken from the REMBRANDT dataset and was parsed through to collect data specifically related to GBM (Glioblastoma). Original data file should also be attached in github labeled 'clinical_2014-01-16.xlsx'

```python
data = pd.read_csv('GBM_data.csv')
avgs = [9.9,7.3,12.4,34.12]
data
#right side of data was calculated inside of excel.
```

```python
dataset = {'no treatment':9.9, 'Ys Yac Yer':34.12, 'Ns Nac Yer':12.4, 'Ns Yac Yer':7.3}
courses = list(dataset.keys())
values = list(dataset.values())
fig = plt.figure(figsize = (10, 5))
plt.bar(courses, values, color ='maroon',width = 0.4)
plt.xlabel("types of treatment")
plt.ylabel("lifespan (months)")
plt.title("different methods of GBM treatment with duration of lifespan")
plt.show()
```

## Caption
Shown above is a bar graph showing the average lifespan of patients after recieving certain forms of treament, the y axis shows the lifespan in months of the patients and the x axis shows the different forms of treatment. these were average values calculated from the dataset. having full titles on the x axis would not work so, the first letter refers to wether or not they recieved it, s referring to steroids, ac referring to anti convulsants and er reffering to EBRT radiation.
