```python
import numpy as np
import math

import scipy as sc
import scipy.constants as con
import scipy.integrate as iii

import matplotlib 
import matplotlib.pyplot as plt
```

```python
def straight(x):
    return -0.5*x+4
def parabola(x):
    return -0.29*x*x-x+12.5
def more(x):
    return 10*(math.e**(-x**2))*(x+1)+1
```

```python
deltax = float(input("please input rectangle width: "))
dataset = np.arange(-5.0,5.00000001-deltax,deltax)
dataset = dataset+0.5*deltax
```

```python
def integ(x):
    #straight line section
    s_res = straight(x)
    s_integ = sum(s_res*deltax)
    s_real = iii.quadrature(straight,-5,5)[0]
    s_diff = 100*(abs(s_real-s_integ)/((s_real+s_integ)/2))
    print('area of straight line is: ', s_integ, 'percentage difference of', s_diff)
    
    #parabola section
    p_res = parabola(x)
    p_integ = sum(p_res*deltax)
    p_real = iii.quadrature(parabola,-5,5)[0]
    p_diff = 100*(abs(p_real-p_integ)/((p_real+p_integ)/2))
    print('area of parabola line is: ', p_integ, 'percentage difference of', p_diff)
    
    #more complicated section
    m_res = more(x)
    m_integ = sum(m_res*deltax)
    m_real = iii.quadrature(more,-5,5)[0]
    m_diff = 100*(abs(m_real-m_integ)/((m_real+m_integ)/2))
    print('area of more complicated line is: ', m_integ, 'percentage difference of', m_diff)
    
    
    
```

```python
integ(dataset)
```
